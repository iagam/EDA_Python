## 1. Number of incidents per State per year.

## 2. Which month's wife's have more tendency to hurt husband?

## 5. Number of incident based on family relationship(ex: wife killed by husband etc.)

```python
family = ['Brother','Sister','Mother','Son','Daughter','Husband','Wife','Father']
fam_incd = homicide[homicide['Relationship'].isin(family)]
fam_incd.head()
```

```python
def desc(df):
    if ((df['Relationship'] == 'Brother') & (df['Perpetrator Sex'] == 'Male')):
            return "Brother killed by brother"
    elif ((df['Relationship'] == 'Brother') & (df['Perpetrator Sex'] == 'Female')):
            return "Brother killed by sister"
    elif ((df['Relationship'] == 'Sister') & (df['Perpetrator Sex'] == 'Male')):
            return "Sister killed by brother"
    elif ((df['Relationship'] == 'Sister') & (df['Perpetrator Sex'] == 'Female')):
            return "Sister killed by sister"
    elif ((df['Relationship'] == 'Mother') & (df['Perpetrator Sex'] == 'Male')):
            return "Mother killed by son"
    elif ((df['Relationship'] == 'Mother') & (df['Perpetrator Sex'] == 'Female')):
            return "Mother killed by daughter"
    elif ((df['Relationship'] == 'Father') & (df['Perpetrator Sex'] == 'Male')):
            return "Father killed by son"
    elif ((df['Relationship'] == 'Father') & (df['Perpetrator Sex'] == 'Female')):
            return "Father killed by daughter"
    elif ((df['Relationship'] == 'Son') & (df['Perpetrator Sex'] == 'Male')):
            return "Son killed by father"
    elif ((df['Relationship'] == 'Son') & (df['Perpetrator Sex'] == 'Female')):
            return "Son killed by mother"
    elif ((df['Relationship'] == 'Daughter') & (df['Perpetrator Sex'] == 'Male')):
            return "Daughter killed by father"
    elif ((df['Relationship'] == 'Daughter') & (df['Perpetrator Sex'] == 'Female')):
            return "Daughter killed by mother"
    elif ((df['Relationship'] == 'Husband') & (df['Perpetrator Sex'] == 'Female')):
            return "Husband killed by wife"
    elif ((df['Relationship'] == 'Wife') & (df['Perpetrator Sex'] == 'Male')):
            return "Wife killed by husband"
    else :
        return "unknown"
```

```python
res = pd.DataFrame()
for i,data in fam_incd.iterrows():
    cur_row = { 
        "Record ID" : data['Record ID'],
        "Incident description" :desc(data)
    }
    res = res.append(cur_row,ignore_index=True)
fam_incd = fam_incd.merge(res)  
fam_incd.head()
```

```python
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from matplotlib import pyplot as plt
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
words = set(stopwords.words('english'))
words
```

```python
amazon = pd.read_excel("D:/Python Files/EDA/Data_sets/Amazon.xlsx")
text = amazon[['reviews.text']]
print(text['reviews.text'][0])
text.head()
```

```python
## Step-1 & Step-2 & Step-3 : Converting everything into lower case and replacing the special charater

text['clear_text'] = text['reviews.text'].str.lower().str.replace('[^a-z ]','')
text.head()
```

```python
## Step-4 : Removing the stopwords

stop = stopwords.words('english')

def sw(text):
    text = [word for word in text.split() if word not in stop]
    return " ".join(text)

## Applying the sw on clear_text columns

text['text_without_stop'] = text['clear_text'].apply(sw)
print(text['clear_text'][0])
print(text['text_without_stop'][0])
```

```python
## Starting the count vector

vec = CountVectorizer()

vec.fit(text['text_without_stop'])
```

```python
## DTM Matrix creation

x = vec.fit_transform(text['text_without_stop'])
x
```

```python
## Convert this DTM into dataframe

DTM = pd.DataFrame(x.toarray(), columns = vec.get_feature_names())

DTM.head()
```

```python
# Calculating the word freq

word_freq = DTM.sum()
word_freq.sort_values(ascending = False).head(3)
```
